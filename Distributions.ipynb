```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import h5py
import numpy as np
```

```python
h5f_ttbb = h5py.File('ttbb.h5','r')
h5f_ttH = h5py.File('ttH.h5','r')
features = ['averageDRbb','maxSumPtDRbb','maxDEta','minDRmbb','minDRmjj','hadHT']
```

```python
for feat in features:
    ttbb = np.array(h5f_ttbb[feat])
    ttH = np.array(h5f_ttH[feat])
    plt.hist(ttbb, histtype='stepfilled', alpha=0.8, normed=True, label='ttbb')
    plt.hist(ttH,histtype='stepfilled', alpha=0.8, normed=True, label='ttH')
    plt.xlabel(feat)
    plt.legend()
    plt.show()
```

### Event Selection
- at least 6 jets with $p_T$>25 GeV
- at least 2 b-jets (highest energy parton within dR<0.4 with PDG ID=5/-5)
